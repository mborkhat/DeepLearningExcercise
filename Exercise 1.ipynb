```python
data=pd.read_csv("https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DA0101EN/auto.csv",names=headers)
```

```python
data.isnull().sum()
```

```python
#e
data = data.replace('?',np.nan) 
```

```python
data.head()
```

```python
#f
data.isnull().sum()[data.isnull().sum()>0]
```

```python
#g
data= data.dropna(subset=['price'])
```

```python
#h and k
mean_col = ['normalized-losses', 'bore', 'stroke', 'horsepower', 'peak-rpm']
data[mean_col] = data[mean_col].astype('float64')
data[mean_col] = data[mean_col].fillna(data[mean_col].mean())
data[mean_col].isnull().sum()
```

```python
#i
data['num-of-doors'].fillna(data['num-of-doors'].mode()[0], inplace=True)
data.head()
```

```python
#j
data.reset_index(inplace=True)
data.head()
```

```python
data.dtypes
```

```python
#k for price
data['price'] = data['price'].astype('int')
data.head()
```

```python
    #L
    data['city-L/100km'] = 235 / data['city-mpg']
    data['highway-L/100km'] = 235 / data['highway-mpg']
    data.head()
```

```python
#M
data['length']=data['length']/data['length'].max()
data['width']=data['width']/data['width'].max()
data['height']=data['height']/data['height'].max()
data.head()
```
