```python
#23
from scipy.stats import pearsonr
corr, _ = pearsonr(data['wheel-base'],data['price'])
print('Pearsons correlation: %.3f' % corr)
```

p-value <0.5 means result is significant else not.

```python
#24
from scipy.stats import f_oneway
f_oneway(data['price'][data['drive-wheels'] == '4wd'],
               data['price'][data['drive-wheels'] == 'fwd'],
               data['price'][data['drive-wheels'] == 'rwd'])
```

p-value is <0.05, hence we do not accept null hypothesis i.e both are correlated and the result is significant .
